```python
from google.colab import files
uploaded = files.upload()
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("sales_dataset.csv")
df.head()
df.info()
df.describe()
print("Total Sales:", df["Sales"].sum())
print(df.groupby("Product")["Sales"].sum())
df.groupby("Date")["Sales"].sum().plot(figsize=(10,5))
plt.title("Daily Sales Trend")
plt.ylabel("Sales")
plt.show()
df.groupby("Product")["Sales"].sum().plot(kind="bar", color="skyblue")
plt.title("Total Sales by Product")
plt.ylabel("Sales")
plt.show()
df.groupby("Region")["Sales"].sum().plot(kind="bar", color="orange")
plt.title("Total Sales by Region")
plt.ylabel("Sales")
plt.show()
df.groupby("Promotion")["Sales"].sum().plot(kind="bar", color=["green","red"])
plt.title("Sales with vs without Promotion")
plt.ylabel("Sales")
plt.show()
```
